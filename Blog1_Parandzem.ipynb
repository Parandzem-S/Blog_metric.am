<h2> Matplotlib (Python) Basic Plots </h2>

Let's see what story we can tell about Women's NBA players' on-court performance in the 2016-2017 season using <i>Matplotlib (Python) library</i> as our visualisation tool. <br> The data is taken from <a><a href= "https://www.kaggle.com/jinxbe/wnba-player-stats-2017/home">www.kaggle.com</a>.

<h3> Start Coding</h3> <br>
The <a> <a href= "https://jupyter.org/"> Jupyter Notebook</a> is used to produce our graphs by using Matplotlib library. Before starting with the first graph, we should do the following:
- Pyplot module from Matplotlib should be imported by using <code>import matplotlib.pyplot as plt</code>.
- <a><a href="https://pandas.pydata.org/">Pandas</a> and other libraries which may be necessary during analysis should be imported as well.  
- The data of your interest should be read using <code>pandas.read()</code> function.

```python
#importing libraries
import matplotlib.pyplot as plt
import pandas as pd

#reading data
my_data=pd.read_csv("https://raw.githubusercontent.com/Parandzem-S/Blog_metric.am/master/Data/WNBA%20Stats.csv")
```

<h3> Histogram, Matplotlib</h3> <br>
First, let's observe players' profiles. For this reason, the first variable chosen to be visualise is players' age. <br>

```python
#histogram to be used to show the distribution of players' age
plt.hist(my_data.Age)
plt.show()
```

Let's do customization of the histogram by giving titles to the axes, changing colors and predefining the number of bins. 

```python
plt.hist(my_data.Age, 
         bins=5, #defining number of bins
         color="#cc0011" #changing color
)
#giving a title to the graph
plt.title("2016-2017 WNBA players' age distribution")

#giving title to y axis
plt.ylabel("Frequencies")

#giving title to x axis
plt.xlabel("Age") 
plt.show()
```

The histogram tells us that the most frequent value was about 25, this means that more than 40 NBA players out of 143 were about 25 years old in 2016-2017 season.<br>

<h3> Box Plot, Matplotlib </h3>

Let's visualise WNBA players' height distribution as well using another plot: <b>box plot</b>. Box plot uses five measures to describe how the data is distributed: <i>minimum, first quartile, median, third quartile, and maximum</i>.

```python
#let's plot simple box plot
plt.boxplot(my_data.Height)
plt.show()
```

To understand the plot better and see 5 measures on it, let's customise the simple box plot:
- let's annotate name of each measure on the plot,
- let's give titles to the graph and axes,
- let's change overall design of boxes by using <code>patch_artist=True</code> and <code>notch=True</code>,
- let's use <code>plt.figure() </code> to change figure size,

```python
import numpy as np
np.round(np.median(my_data.Height),0)
```

```python
#making the plot bigger
plt.figure(figsize=(8,6))

#creating the plot
plt.boxplot(my_data.Height,
            patch_artist=True,  #changing face color
            notch=True #changing a rectangular form to a notched form
           )
#annotating median value
plt.annotate("%0.0F"% np.median(my_data.Height), xy=(1, 186), xytext=(1,186))

#annotating names of 5 measures mensioned above
plt.annotate('Minimum', xy=(1, 165), xytext=(1.1,164))
plt.annotate('First Quartile', xy=(1, 176), xytext=(1.1,176))
plt.annotate('Median', xy=(1, 184), xytext=(1.1,184))
plt.annotate('Third Quartile', xy=(1, 190), xytext=(1.1,190))
plt.annotate('Maximum', xy=(1, 205), xytext=(1.1,205))

#giving titles
plt.title("WNBA's Players Height (cm) Distribution")
plt.show()
```

<b> So far, we can conclude that majority of WNBA player's were mostly 24-28 years old and half of them had more than 185 cm height. </b>

<h3> Bar chart, Matplotlib </h3>

Let's observe the realtionship between the positions they played in and their height by using bar plot: <code> plt.bar()</code>. 

```python
plt.figure(figsize=((8,6)))
#draw line
plt.bar(my_data.Pos, my_data.Height, color="#cc0011")

#lets' define the range of y axis
plt.ylim(0,250)

#giving titles
plt.title("WNBA Players Height by Position")
plt.ylabel("Height (cm)")
plt.xlabel("Positions")

plt.show()
```

We can see from the plot that players who played in the center (C position) had relatively higher heights. And shortest ones played as guards. <br>
Let's use horizontal bar chart to see whther there is difference in players' weights by position.

```python
plt.figure(figsize=((8,6)))

#draw plot
plt.barh(my_data.Pos, my_data.Weight,color="#5394ef")

#giving titles
plt.title("WNBA Players Weight by Position")
plt.ylabel("Positions")
plt.xlabel("Weight (kg)")
plt.xlim(0,130)

plt.show()
```

As it was expected those who played in center have the highest weight. 

<b>Therefore, we can conlcude that the center players are the tallest and heaviest ones, and the guards are the shortest and lightest ones.</b>

<h3> Scatter Plot, Matplotlib </h3>

Lets see whether WNBA players performance (Field Goals Made - FGM) has any relationship with their body mass index (<a><a href= "http://www.nba.com/nbafit/kid/nutrition/weight/30219_body_mass_index_bmi.html">BMI</a>).

```python
plt.figure(figsize=((8,6)))

#draw plot
plt.scatter(my_data.Height,my_data.FGM,color="#072659")

#giving titles
plt.title("Relationship between BMI and FGM")
plt.ylabel("FGM")
plt.xlabel("BMI")

plt.show()
```

No obvious relationship exists between height and performace. Let's visualise FGM with games played. 

```python
plt.figure(figsize=((8,6)))

#draw plot
plt.scatter(my_data["Games Played"],my_data.FGM,color="#63001e")

#giving titles
plt.title("Relationship between BMI and FGM")
plt.ylabel("FGM (Field Goals Made)")
plt.xlabel("Games played")

plt.show()
```

It can be claerly indicated from the plot above that there is a relationship between players performance and games played, however the relationship is not linear but most probably it's more exponential.

To summarise, WNBA players are mostly 24-28 years old, the center players are the tallest and heaviest ones. In the first 25 games the number of field goals made increases slowly, while after that the number of goals scored increases exponencially.
